```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population=np.arange(5,100,5)
population
```

```python
#population mean
mean_pop=sum(population)/len(population)
mean_pop
```

```python
#how far the points are from the mean 
sq_pop_SD=[(i-mean_pop)**2 for i in population]
sq_pop_SD
```

```python
pop_varience=sum(sq_pop_SD)
pop_varience
```

```python
pop_varience/len(population)
```

## Sample

```python
sample=np.arange(5,40,5)
sample
```

```python
smean=sum(sample)/len(sample)
smean
```

```python
sq_sSd=[(i-smean)**2 for i in sample]
```

```python
s_varience=sum(sq_sSd)
s_varience
```

```python
sample_varience=s_varience/(len(sample)-1)
round(sample_varience,2)
```

## Measure Of Dispersion

```python
#sample SD
import math
std_dev=math.sqrt(sample_varience)
std_dev
```
