## ploating and the Visulization of the data

```python
plt.figure(figsize=(12,7))

sns.scatterplot(data=df,x='healthy_eating',y='active_lifestyle',color='blue',edgecolor='linen',alpha=0.5)
plt.title("healthy_eating vs active_lifestyle Scatterplot")
plt.xlabel("healthy_eating")
plt.ylabel("active_lifestyle")
plt.show()
```

```python
plt.figure(figsize=(12,7))

sns.scatterplot(data=df,x='age',y='salary',color='red',edgecolor='linen',alpha=0.5)
plt.title("Age vs Salary Scatterplot")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(12,7))

sns.scatterplot(data=df,x='healthy_eating',y='salary',color='green',edgecolor='linen',alpha=0.5)
plt.title("healthy_eating vs Salary Scatterplot")
plt.xlabel("healthy_eating")
plt.ylabel("Salary")
plt.show()
```

```python
# countplot of the groups
plt.figure(figsize=(12,7))
df['groups'].value_counts().plot(kind='bar')

plt.title("Contplot of the Blood Group")
plt.xlabel("Groups")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,7))

sns.histplot(data=df,x='salary',color='blue',edgecolor='black',bins=10)
plt.title("Histogram of the Salary")
plt.ylabel("Salary")
plt.xlabel("Bins")
plt.show()
```
