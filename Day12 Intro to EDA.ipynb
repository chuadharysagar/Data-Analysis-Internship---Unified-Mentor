### OUTLIERS IQR Inter Quartalile Range

```python
#first quatile 25% #25 percentage mean is Q1
Q1=np.percentile(data['Unit Cost'],25)
Q1
```

```python
Q3=np.percentile(data['Unit Cost'],75)
Q3
```

```python
IQR= Q3-Q1
IQR
```

```python
low=Q1-1.5*IQR
upper= Q3+1.5*IQR
```

```python
#empty list for storing ouliers
outliers=[]
```

```python
# values greater then Q# and Lower than the Q1 store in list 
for i in data['Unit Cost']:
    if((i>upper) or (i<low)):
        outliers.append(i)
```

```python
#no outliers in data
outliers
```

```python
#now to remove the outliers from the lsit 
# all the not outlier values
for i in data['Unit Cost']:
    if i not in outliers:
        print(i)
```

```python
#store in the list non outlier values
nonoutliesre=[i for i in data['Unit Cost'] if i not in outliers]
```
