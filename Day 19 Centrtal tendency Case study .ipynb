```python
#distribution of the data and various features
# create the histogram of the fixed acid columns 
plt.figure(figsize=(11,6))

sns.histplot(data=df,x='fixed acidity',color='pink',edgecolor='red',alpha=0.5,bins=5)

plt.title("Histopgram of fixed acidity")
plt.ylabel("count")
plt.xlabel("fixed acidity")
plt.show()
```

```python
#Osevarvastion 
# 1. the data is noramally distributated
# 2. th maxim value count of the fixed acid column is between 6 to 8
```

```python
3 #now then we have to see the central tendency mean , median and mode of the data
```

```python
#calculate mean 
round(df['fixed acidity'].mean(),2)
```

```python
#calculate late the mode
df['fixed acidity'].median()
```

```python
#plot hte histogram and then show the mean and the median of the that colunm
plt.figure(figsize=(11,6))

sns.histplot(data=df,x='fixed acidity',color='blue',edgecolor='red',alpha=0.5,bins=5)
plt.title("Histogram with the Mean and median")
plt.xlabel("fixed acidity")
plt.ylabel("Count")

#vertical lines
plt.vlines(df['fixed acidity'].mean(),ymin=0,ymax=4000,colors="black",label="mean")
plt.vlines(df['fixed acidity'].median(),ymin=0,ymax=4000,colors="yellow",label="median")

plt.legend()
plt.show()
```

```python
# Observation 2
# the mean and the median are very close to each other 
# we can use either of it 
```

```python
# Create the histogram for the volatile acidity 
plt.figure(figsize=(11,6))

sns.histplot(data=df,x=df['volatile acidity'],color='orange',edgecolor='black',alpha=0.5,
            bins=5)
plt.title("Histo Graam of volatile acidity")
plt.xlabel("volatile acidity")
plt.ylabel("frequency Count")
plt.show()

#skwwed data 
```

```python
# Obsevation the data is not distributed uniforamlly 
plt.figure(figsize=(11,6))

sns.distplot(x=df['volatile acidity'],color='blue')
plt.title("Histo Graam of volatile acidity")
plt.xlabel("volatile acidity")
plt.ylabel("frequency Count")
plt.show()

#skwwed data 
```

```python
# Calculate the Skewness of the function 
df['volatile acidity'].skew()
```

```python
#plot hte histogram and then show the mean and the median of the that colunm
plt.figure(figsize=(11,6))

sns.histplot(data=df,x='volatile acidity',color='green',edgecolor='red',alpha=0.5,bins=5)
plt.title("Histogram with the Mean and median")
plt.xlabel("fixed acidity")
plt.ylabel("Count")

#vertical lines
plt.vlines(df['volatile acidity'].mean(),ymin=0,ymax=4000,colors="black",label="mean")
plt.vlines(df['volatile acidity'].median(),ymin=0,ymax=4000,colors="yellow",label="median")

plt.legend()
plt.show()
```

```python
#Citric acid columns analysis 
df['citric acid'].median()
```

```python
#histogram of the citric acid column and the central tendency distributin 
plt.figure(figsize=(11,6))

sns.histplot(data=df,x= df['citric acid'], color='blue',edgecolor='gold',alpha=0.5,bins=5)
plt.title("Histpgram of the citric acid columns and central tendecy")
plt.xlabel("citric acid")
plt.ylabel("count_number")

# vertical lines for the mean and the median of the data
plt.vlines(df['citric acid'].mean(),ymin=0,ymax=3000,color="red",label="Mean")
plt.vlines(df['citric acid'].median(),ymin=0,ymax=3000,color="pink",label="Median")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11,6))

sns.distplot(x=df['citric acid'],color='blue')
plt.title("Histo Graam of citric acid")
plt.xlabel("citric acid")
plt.ylabel("frequency Count")
plt.show()
```

```python
# count plot of the quality columns from the dataset
plt.figure(figsize=(11,6))

df['quality'].value_counts().plot(kind='bar')
plt.title("Quqlity column count plot")
plt.xlabel("quality")
plt.ylabel("count")
plt.show()
```
